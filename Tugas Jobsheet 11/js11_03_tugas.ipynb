## TUGAS 1

```python
import numpy as np
import pandas as pd

df = pd.read_csv('data/wbc.csv')

# slice kolom terakhir --> Unnamed column
df = df.iloc[:,:-1]

df.head()
```

```python
X = df.iloc[:, 2:]
y = df['diagnosis']

X.head()
```

```python
from sklearn.model_selection import RepeatedKFold

cv = RepeatedKFold(n_splits=4, n_repeats=3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

dt = DecisionTreeClassifier()

params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': list(range(5,11))
}

grid = GridSearchCV(dt, param_grid=params, cv=cv)

%timeit grid.fit(X, y)

score = grid.score(X,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {grid.best_params_}')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV

clf = RandomizedSearchCV(dt, param_distributions=params, cv=cv)
%timeit grid.fit(X, y)
search = clf.fit(X, y)

nilai = clf.score(X,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {search.best_params_}')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
from sklearn import svm

svc = svm.SVC()
params = {
    'kernel': ['rbf', 'linear', 'poly'],
    'C': [3],
    'degree': [1]
}

random = RandomizedSearchCV(svc, param_distributions=params, cv=cv, n_jobs=-1, )
%timeit random.fit(X, y)

score = random.score(X,y)

print(f'Hasil evaluasi Random Search: {score}')
print(f'Konfigurasi hyperparameter Random Search: {random.best_params_}')
```

```python
from sklearn.model_selection import GridSearchCV
params = {
    'kernel': ['rbf', 'linear', 'poly'],
    'C': [3],
    'degree': [1]
}

grid = GridSearchCV(svc, param_grid=params, cv=cv, n_jobs=-1)

%timeit grid.fit(X, y)
score = grid.score(X,y)

print(f'Hasil evaluasi Grid Search: {score}')
print(f'Konfigurasi hyperparameter Grid Search: {grid.best_params_}')
```
