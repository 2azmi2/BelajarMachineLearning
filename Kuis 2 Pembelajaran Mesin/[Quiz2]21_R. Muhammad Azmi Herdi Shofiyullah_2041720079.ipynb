```python
#import fungsi yang diperlukan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

#Membaca data dari file excel
df = pd.read_excel('data/dataset_quiz.xlsx')

df.head(5)
```

```python
#Mengecek isi dari atribut keterlibatan_vltr untuk menentukan apakah atribut ini adalah label (clue berada di soal no 4)
print(df['keterlibatan_vltr'].value_counts())
print('\n')
```

## Menentukan Label
Menurut informasi yang didapatkan pada soal 4 dimana diwajibkannya atribut keterlibatan volunteer maka saya berkesimpulan bahwa atribut keterlibatan volunteer adalah label yang akan digunakan. Selain itu dengan melakukan pengecekan apa saja isi dari atribut keterlibatan_vltr hanya memiliki 2 hasil maka memperkuat kesimpulan saya bahwa keterlibatan_vltr adalah label yang akan digunakan.

```python
#Melakukan seleksi data
X = df_copy.iloc[:,6:]
y = df_copy.iloc[:,5]

X.head()
```

```python
# import Split data
from sklearn.model_selection import train_test_split

# Split data training dan dan lainnya
# data lainnya, akan kita split lagi menjadi validasi dan testing.
df_train, df_unseen = train_test_split(df_copy, test_size=0.2, random_state=0)

# Split lagi antara validasi dan testing
df_val, df_test = train_test_split(df_unseen, test_size=0.5, random_state=0)

# Cek masing-masing ukuran data

print(f'Jumlah data asli: {df.shape[0]}')
print(f'Jumlah data train: {df_train.shape[0]}')
print(f'Jumlah data validasi: {df_val.shape[0]}')
print(f'Jumlah data test: {df_test.shape[0]}')

# Cek rasio tiap label
print('=========')
print(f'Jumlah label data asli:\n{df_copy.keterlibatan_vltr.value_counts()}')
print(f'Jumlah label data train:\n{df_train.keterlibatan_vltr.value_counts()}')
print(f'Jumlah label data validasi:\n{df_val.keterlibatan_vltr.value_counts()}')
print(f'Jumlah label data test:\n{df_test.keterlibatan_vltr.value_counts()}')
```

```python
#Melakukan split train
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
# Instansiasi Decision Tree Classifier
dt = DecisionTreeClassifier(random_state=1)

# Sesuaikan dt ke set training
dt.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

# Memprediksi test set labels
y_pred = dt.predict(X_test)

#  menghitung set accuracy
acc = accuracy_score(y_test, y_pred)
print(f"Test set accuracy: {acc}")
```

```python
# Memanggil fungsi DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

# Instansiasi dt_entropy, set 'entropy' sebagai kriteria informasi
dt_entropy = DecisionTreeClassifier(max_depth=8, criterion='entropy', random_state=1)

# Fit dt_entropy kedalam training set
dt_entropy.fit(X_train, y_train)
```

```python
# Instansiasi dt_gini, set 'gini' sebagai kriteria informasi
dt_gini = DecisionTreeClassifier(max_depth=8, criterion='gini', random_state=1)

# Fit dt_gini ke training set
dt_gini.fit(X_train, y_train)
```

```python
#memanggil fungsi Accuracy Score
from sklearn.metrics import accuracy_score

# menggunakan dt_entropy untuk memprediksi test set labels
y_full_pred_entropy = dt_entropy.predict(X_test)
y_full_pred_gini = dt_gini.predict(X_test)

# mengevaluasi accuracy_entropy
accuracy_entropy = accuracy_score(y_test, y_full_pred_entropy)
accuracy_gini = accuracy_score(y_test, y_full_pred_gini)

# Print accuracy_entropy
print("Accuracy achieved by using entropy: ", accuracy_entropy)

# Print accuracy_gini
print("Accuracy achieved by using gini: ", accuracy_gini)
```

## Naive Bayes

```python
#Memanggil fungsi Multinomial Naive Bayes 
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Inisiasi MultinomialNB
mnb = MultinomialNB()

# Fit model
mnb.fit(X_train, y_train)

# Prediksi dengan data training
y_pred_train = mnb.predict(X_train)

# Evaluasi akurasi data training
acc_train = accuracy_score(y_train, y_pred_train)

# Prediksi dengan data training
y_pred_test = mnb.predict(X_test)

# Evaluasi akurasi data training
acc_test = accuracy_score(y_test, y_pred_test)

# Print hasil evaluasi
print(f'Hasil akurasi data train: {acc_train}')
print(f'Hasil akurasi data test: {acc_test}')
```

## Hyperparameter

```python
#Memanggil fungsi Repeated K Fold
from sklearn.model_selection import RepeatedKFold

# instansiasi cv dengan menggunakan ReapetedKFold
cv = RepeatedKFold(n_splits=4, n_repeats=3, random_state=42)
```

```python
#import fungsi DT dan Grid Search
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# inisiasi model
dt = DecisionTreeClassifier()

# Definisikan hyperparameter yang akan digunakan
# Menggunakan 2 kriteria yang sebelumnya telah digunakan ('entropy' dan 'gini' pada DT)
params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': list(range(5,11))
}

# inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
grid = GridSearchCV(dt, param_grid=params, cv=cv)

# Fit / latih berdasarkan grid
# Menghitung waktu komputasi dengan %timeit
%timeit grid.fit(X, y)

# Evaluasi dengan score
score = grid.score(X,y)

# Cetak hasil evaluasi dan konfigurasi terbaik
print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter: {grid.best_params_}')
```

## Random Search

```python
# import random search
from sklearn.model_selection import RandomizedSearchCV

# instansiasi clv dengan random search
clf = RandomizedSearchCV(dt, param_distributions=params, cv=cv)

# Fit / latih berdasarkan grid
# Menghitung waktu komputasi dengan %timeit
%timeit grid.fit(X, y)

# Fit dengan menggunakan random search
search = clf.fit(X, y)

# Cari nilai akurasi dengan menggunakan score
nilai = clf.score(X,y)

#tampilkan hasil akurasi dan konfigurasi terbaik
print(f'Hasil evaluasi: {nilai}')
print(f'Konfigurasi hyperparameter: {search.best_params_}')
```
